```python
from rdflib import Graph
from rdflib import URIRef, BNode, Literal
from rdflib import Namespace
from rdflib.namespace import OWL, RDF, RDFS, FOAF, XSD

import pandas as pd
```

```python
df = pd.read_csv("stats.csv")
df
```

```python
df.head(1)
```

```python
def createTriples():
   
    g = Graph()
    #g.parse("talhaking.owl", format="ttl")            
            
    #Special namspaces to create  
    dbo = Namespace("https://dbpedia.org/ontology/")
    dbp = Namespace("https://dbpedia.org/property/")
    dbr = Namespace("https://dbpedia.org/page/")
    pl = Namespace("http://www.sports.com/sports/")
    
    #Prefixes
    g.bind("foaf", FOAF) #FOAF is given as defaulty namespace
    g.bind("dbo", dbo) #dbo is a newly created namespace 
    g.bind("dbp", dbp)
    g.bind("pl", pl)
    
     # Links to External Datasets/Ontologies
    
    g.add((pl.Player, RDFS.subClassOf, dbo.athlete ))
    g.add((pl.Player, OWL.sameAs, dbo.SoccerPlayer))
    g.add((pl.Club, OWL.sameAs, dbo.team))
    
    
    for i in range(len(df.head(100))):
        
        Club = df.iloc[i,0]
        Wins = df.iloc[i,1]
        Losses = df.iloc[i,2]
        Goals = df.iloc[i,3]
        YellowCard = df.iloc[i,4]
        RedCard = df.iloc[i,5]
        HitWoodworK = df.iloc[i,8]
        Touches = df.iloc[i,34]
        Seasons = df.iloc[i,41]
        
        #######################################################
        
        Club = Club.replace(' ', '_') 
        Club = Literal(Club, datatype=XSD.string)     
        
        str1 = 'http://www.sports.com/sports/' + Club        
        team = URIRef(str1)
        
        g.add((team, RDF.type, pl.Club))    
        
        str2 = 'https://dbpedia.org/page/' + Club        
        dbr = URIRef(str2)
        g.add((team, OWL.sameAs, dbr))
        
        #######################################################
        
        Wins = Literal(Wins, datatype=XSD.int)
        g.add((team, pl.hasWins,Wins))
        
        #######################################################
        
        Losses = Literal(Losses, datatype=XSD.int)
        g.add((team, pl.hasLosses,Losses))
        
        #######################################################
        
        Goals = Literal(Goals, datatype=XSD.int)
        g.add((team, pl.scoresGoal,Goals))
        
        #######################################################
        
        YellowCard= Literal(YellowCard, datatype=XSD.int)
        g.add((team, pl.getsYellow,YellowCard))
        
        #######################################################
        
        
        RedCard= Literal(RedCard, datatype=XSD.int)
        g.add((team, pl.getsRed,RedCard))
        
        #######################################################
        
        Touches= Literal(Touches, datatype=XSD.int)
        g.add((team, pl.hasTouches,Touches))

        #######################################################
        
        Seasons = Literal(Seasons, datatype=XSD.string)
        g.add((team, pl.playsMatch,Seasons))
        
    print("\nSaving graph to 'newGraph.ttl':\n\n")
    g.serialize(destination='newGraph.ttl', format='ttl')
    print(g.serialize(format="ttl"))    
           

createTriples()
    
```
